*Для каждой книги посчитаем количество обзоров и среднюю оценку:*

````python
query = """
 SELECT
   ratings.book_id,
   AVG(rating) AS avg_rating,
   subq.reviews_count AS reviews_count
 FROM 
   ratings
 LEFT JOIN
  (
   SELECT
    reviews.book_id,
    COUNT(reviews.review_id) AS reviews_count
   FROM
    reviews
   GROUP BY
    reviews.book_id
  ) AS subq 
 ON 
   ratings.book_id = subq.book_id
 WHERE
   subq.book_id IS NOT NULL
 GROUP BY
    ratings.book_id,
    subq.reviews_count
 ORDER BY
    avg_rating DESC;
"""

pd.io.sql.read_sql(query, con = engine)
````

````python
query = """
 SELECT
   publishers.publisher_id,
   publisher,
   MAX(subq.books_count) AS max_publishes
 FROM 
   publishers
 LEFT JOIN
  (
  SELECT 
    books.publisher_id,
    COUNT(books.book_id) AS books_count
  FROM books
  WHERE books.num_pages > 50
  GROUP BY
    books.publisher_id
  ) AS subq 
 ON 
   publishers.publisher_id = subq.publisher_id
 WHERE
   subq.publisher_id IS NOT NULL
 GROUP BY
    publishers.publisher_id
 ORDER BY
     max_publishes DESC
 LIMIT 
     1;
"""

pd.io.sql.read_sql(query, con = engine)
````

*Определим автора с самой высокой средней оценкой книг (учитывайте только книги с 50 и более оценками):*

````python
query = """
 SELECT
    authors.author AS author,
    AVG(subq.avg_rating) AS avg_author_rating
 FROM
    authors
 INNER JOIN
    books ON books.author_id = authors.author_id
 INNER JOIN
   ( 
    SELECT
      book_id,
      AVG(rating) AS avg_rating,
      COUNT(rating_id) AS ratings_count
    FROM
      ratings
    GROUP BY
      book_id
    ) AS subq 
    ON subq.book_id = books.book_id
    WHERE
      subq.ratings_count > 50
  GROUP BY
    author
  ORDER BY
    avg_author_rating DESC
   LIMIT 
     1;  
   """

pd.io.sql.read_sql(query, con = engine)
````

Я хотел подтолкнуть тебя к более простому и оптимальному варианту. Эту задачу можно было решить без джойнов. 
    
1. Первым шагом мы фильтруем пользователей (```query_rev_01```)
    
2. В reviews есть и username и review_id, просто потом фильтруем пользователей через ```where reviews.username in (*твой подзапрос*)``` (```query_rev_02```)

````python
# КОД РЕВЬЮЕРА

query_rev_01 = (
'''
    select username
    from ratings
    group by username
    having count(rating_id) > 50
    
''')

pd.io.sql.read_sql(query_rev_01, con = engine)
````

````python
# КОД РЕВЬЮЕРА

query_rev_02 = (
'''
    select reviews.username, count(review_id) as cnt_rev
    from reviews
    where reviews.username in ( select username
                        from ratings
                        group by username
                        having count(rating_id) > 50)
    group by reviews.username
    
''')

pd.io.sql.read_sql(query_rev_02, con = engine)
````

````python
# КОД РЕВЬЮЕРА

query_rev_03 = (
'''
with table_full as 
    ( 
      select reviews.username, count(review_id) as cnt_rev
      from reviews
      where reviews.username in ( select username
                                  from ratings
                                  group by username
                                  having count(rating_id) > 50 )
      group by reviews.username
    )
    
select avg(cnt_rev)
from table_full
    
''')

pd.io.sql.read_sql(query_rev_03, con = engine)
````
